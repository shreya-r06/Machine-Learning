```python
import pandas as pd
```

```python
df = pd.read_csv("heart_attack_risk.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df.info()
```

```python
df.head()
```

```python
df['Stress_Level']= le.fit_transform(df['Stress_Level'])
df['Chest_Pain_Type']= le.fit_transform(df['Chest_Pain_Type'])
df['Thalassemia']= le.fit_transform(df['Thalassemia'])
df['ECG_Results']= le.fit_transform(df['ECG_Results'])
df['Heart_Attack_Risk']= le.fit_transform(df['Heart_Attack_Risk'])
df['Physical_Activity_Level'] = le.fit_transform(df['Physical_Activity_Level'])
```

```python
df.info()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,accuracy_score
```

```python
X = df.drop('Heart_Attack_Risk',axis=1)
y = df['Heart_Attack_Risk']
X_train,x_test,Y_train,y_test = train_test_split(X,y,test_size =0.3,random_state=42)
```

```python
import matplotlib.pyplot as plt

acc=[]
for i in range(1,26):
    classifier = KNeighborsClassifier(n_neighbors = i,algorithm = 'auto',
                             metric='cityblock')
    classifier.fit(X_train,Y_train)
    y_pred = classifier.predict(x_test)
    acc.append(1-accuracy_score(y_pred,y_test))
```

```python
plt.figure(figsize=(15,6))
plt.plot(range(1,26),acc,marker='o')
plt.xlabel('K value')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

k_range = range(1, 21)
cv_scores = [cross_val_score(KNeighborsClassifier(n_neighbors=k), X_train, Y_train, cv=5).mean() for k in k_range]

optimal_k = k_range[np.argmax(cv_scores)]
print(f"Best K: {optimal_k}")
```

```python
distances = ['cityblock','cosine','euclidean']
acc_dist=[]
for i in distances:
    classifier = KNeighborsClassifier(n_neighbors = 20,algorithm = 'auto',
                             metric=i)
    classifier.fit(X_train,Y_train)
    y_pred = classifier.predict(x_test)
    acc_dist.append(accuracy_score(y_pred,y_test))
```
